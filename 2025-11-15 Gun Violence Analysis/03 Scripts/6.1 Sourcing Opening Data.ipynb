```python
#checking for missing values
missing_values = gunv.isnull().sum()
print(missing_values)
```

```python
# These columns have too many missing values to be useful
columns_to_remove = ['participant_relationship', 'location_description', 'participant_name']
data_cleaned = gunv.drop(columns=columns_to_remove)
```

```python
# Dropping participant_age because it's too incomplete to be useful
data_cleaned = data_cleaned.drop(columns=['participant_age'])
```

```python
#checking data set for removal
data_cleaned
```

```python
# If a value is missing, we will label it as "Unknown"
categorical_columns = ['gun_stolen', 'gun_type', 'participant_age_group', 'participant_gender']
for column in categorical_columns:
    data_cleaned[column] = data_cleaned[column].fillna('Unknown')
```

```python
# Fill missing district values with "Unknown" for cleaner visuals
district_cols = ['congressional_district', 'state_house_district', 'state_senate_district']

for col in district_cols:
    data_cleaned[col] = data_cleaned[col].astype('string').fillna('Unknown')
```

```python
# Converting the date column to a proper datetime format
data_cleaned['date'] = pd.to_datetime(data_cleaned['date'], errors='coerce')
```

```python
# Fill categorical/text columns with "Unknown"
text_columns = ['address', 'incident_characteristics', 'participant_status', 'participant_type']
for col in text_columns:
    data_cleaned[col] = data_cleaned[col].fillna('Unknown')
```
