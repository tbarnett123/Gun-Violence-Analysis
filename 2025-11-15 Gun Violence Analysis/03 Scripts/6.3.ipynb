```python
#creating variables to assist with analysis
#creating total casualties as sum of killed and injured
dcp1['total_casualties'] = dcp1['n_killed'] + dcp1['n_injured']
```

```python
#checking the dataframe to see the columns
print(dcp1[['n_killed', 'n_injured', 'total_casualties']].head())
```

```python
#creating variable if any person was killed in the incident
dcp1['is_fatal'] = (dcp1['n_killed'] > 0).astype(int)

#creating variable if any person was injured
dcp1['has_injuries'] = (dcp1['n_injured'] > 0).astype(int)

# creating variable if total casualties are 4 or more (This can be used as a high severity flag)
dcp1['high_casualty'] = (dcp1['total_casualties'] >= 4).astype(int)

# creating a variable if more than one victim (killed or injured)
dcp1['multi_victim'] = (dcp1['total_casualties'] > 1).astype(int)
```

# 04 Plotting a Choropleth

```python
#creating a df with just the states and those killed to be plotted

data_to_plot = state_subset[['state','n_killed']]
data_to_plot.head()
```

```python
#plotting a chorpleth map

map = folium.Map(location = [37.8,-99], zoom_start = 4)

folium.Choropleth(
    geo_data = country_geo,
    data = data_to_plot,
    columns = ['state', 'n_killed'],
    key_on = 'feature.properties.name',
    fill_color = 'YlOrBr', fill_opacity=0.6,
    line_opacity=0.1,
        legend_name = "Number Killed").add_to (map)

folium.LayerControl().add_to(map)

map
```

The choropleth map reveals strong geographic differences in gun-violence fatalities across the United States. Maine, California, Nevada, and Delaware stand out with the highest fatality counts, shown by the darkest shading on the map. These states experienced the most severe outcomes in the dataset, either due to a higher volume of incidents or a smaller number of high-impact events that drove fatalities upward. Several other states, including Texas, Florida, Tennessee and Georgia, fall into moderate fatality ranges. In contrast, much of the Midwest, Mountain West and northern New England appears very light on the map, indicating little to no fatality data in the filtered subset. 

Overall, the map highlights substantial regional variation in the severity of gun-violence incidents.
